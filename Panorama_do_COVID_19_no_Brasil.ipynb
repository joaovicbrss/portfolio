# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados


Os dados foram obtidos através do [GitHub](https://github.com/owid) do projeto [Our World in Data](https://ourworldindata.org/), e tem esforços colaborativos entre pesquisadores da Universidade de Oxford, que são os editores científicos do conteúdo do site; e a organização sem fins lucrativos [Global Change Data Lab](https://global-change-data-lab.org/) (GCDL), que publica e mantém o site e as ferramentas de dados. A GCDL tem o objetivo de promover a educação sobre como as condições globais de vida e o meio ambiente estão mudando, por meio da produção e manutenção de recursos públicos online.

Os dados são coletados de fontes oficiais de cada país e atualizados diariamente pelo [Center for Systems Science and Engineering (CSSE)](https://github.com/CSSEGISandData/COVID-19) da The Johns Hopkins University. A Our World in Data disponibiliza esses dados de forma estruturada e será atualizado ao longo da duração da pandemia. 

A COVID-19 é uma doença descoberta a pouco tempo, provocando um aumento na dificuldade de contenção da influenza. Um dos desafios também é a dificuldade de levantamento de dados verídicos e confiáveis, pois em alguns países existe um tipo de negligência. Buscando uma diminuição artificial dos números para esses países não sofram fechamento de fronteiras por conta da COVID-19 em seu território.

No caso do Brasil, os dados são atualizados pelo Ministério da Saúde através de informações oficiais repassadas pelas Secretarias Estaduais de Saúde das 27 Unidades Federativas Brasileiras. Os dados fornecidos pelos estados são consolidados e disponibilizados publicamente todos os dias às 19h.

* `people_fully_vaccinated` - *Número Total de Pessoas que Receberam Todas as Doses Prescritas Pelo Protocolo de Vacinação‎*
* `total_boosters` - *Número Total de Doses de Reforço da Vacina da Covid-19 Administradas (Doses Administradas Além do Número Prescrito pelo Protocolo de Vacinação)‎*
* `new_vaccinations` - *Novas Doses de Vacinação Covid-19 Administradas (Calculadas Apenas por Dias Consecutivos)*

* `stringency_index` - *Índice de Restrição de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamentos de escolas, fechamentos de locais de trabalho e proibições de viagens, redimensionado para um valor de 0 a 100 (100 = resposta mais estrita)*
* `population` - *População em 2020*
* `population_density` - *Densidade Populacional*
* `median_age` - *Idade Mediana da população, projeção da ONU para 2020*
* `aged_65_older` - *Parcela da população com 65 anos ou mais, ano mais recente disponível*
* `aged_70_older` - *Parcela da população com 70 anos ou mais em 2015*

* `cardiovasc_death_rate` - *Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas)*

É possível também construir um gráfico de barras mostrando quais países tem mais mortes por COVID-19.

```
#Gráfico de barras para os 5 países com mais mortes
top_total_deaths = df.loc[df.date == '2021-09-21', ['location', 'total_deaths','continent']].sort_values(by="total_deaths", ascending=False)[0:20]
top_total_deaths.dropna(axis=0,inplace=True)

fig, ax = plt.subplots(figsize=(8,6))
ax.set_xlabel('Países', fontsize=20, color='gray')
ax.set_ylabel('Número de Mortes', fontsize=20, color='gray')
ax.set_title('Top 5 - Países Número de Mortos', fontsize=23, loc='center')
ax.bar('location', 'total_deaths', data=top_total_deaths[0:5], color='green')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

def major_tick(x, pos):
  if x < 1000:
    return f'{x:.0f}'
  else:
    return f'{x//1000:.0f}K'
ax.yaxis.set_major_formatter(FuncFormatter(major_tick))


fig.tight_layout()
```

Analisando a evolução da COVID-19 no mundo, através do gráfico de linhas facilita-se a leitura e observação dos dados de casos da doença em relação ao tempo. Fica visível toda a trajetória dos casos no mundo desde o início da pandemia.

```
#Gráfico de linhas para ver a evolução do total de casos de COVID-19 no mundo.

from matplotlib.ticker import FuncFormatter

total_world = df.loc[df.location == 'World',['date','total_cases']]
total_world.date = pd.to_datetime(total_world.date)
total_world.dropna(axis=0,inplace=True)
fig, ax_world = plt.subplots(figsize=(15,6))
ax_world.set_xlabel('Tempo',fontsize=20, color='gray')
ax_world.set_ylabel('Casos',fontsize=20, color='gray')
ax_world.set_title('Evolução de Casos de COVID-19 no Mundo',fontsize=23)
ax_world.spines['right'].set_visible(False)
ax_world.spines['top'].set_visible(False)

ax_world.set_ylim(ymax=250000000)

def major_tick(x, pos):
  if x < 1000000:
    return f'{x:.0f}'
  else:
    return f'{x//1000000:.0f}M'

ax_world.yaxis.set_major_formatter(FuncFormatter(major_tick))

line, = ax_world.plot(total_world.date, total_world.total_cases,lw=3,color='green')

plt.tight_layout()
```

Para ajudar na verificação de outliers, foi utilizado os gráficos de caixa(boxplot) e violino. 


O boxplot é um diagrama que tem o objetivo de demonstrar as medidas estatísticas de um conjunto de dado. Nele é possível ver como está a distribuição entre os quartis, a mediana e os outliers.


O gráfico de violino tem as mesmas características do boxplot mais uma capacidade adicional de verificar a frequência do conjunto de dado.

```
#Verificando a existência de outliers

fig, ax = plt.subplots(nrows=2, figsize=(8,6))

df.gdp_per_capita.plot(kind='box', vert=False, ax=ax[0])
ax[0].set_xlabel('gdp_per_capita', color='gray', size=12)
ax[0].yaxis.set_visible(False)

sns.violinplot(df.gdp_per_capita, ax=ax[1], color='green')
ax[1].set_xlabel('gdp_per_capita', color='gray', size=12)

def major_tick(x, pos):
  if x < 1000:
    return f'{x:.0f}'
  else:
    return f'{x//1000:.0f}K'

ax[0].xaxis.set_major_formatter(FuncFormatter(major_tick))
ax[1].xaxis.set_major_formatter(FuncFormatter(major_tick))


fig.tight_layout()
```

Para ter uma visão da evolução da COVID-19 no Brasil, foi construído um gráfico com visualização linear ou logarítmica. Desta forma, os dados ficam mais ver a curva de casos e mortes da doença. 

Quando utilizar escala Logarítmica e Aritmética (ou linear)? Quando existe uma diferença de ordem de grandeza dos valores, dificultando a visualização dos dados.

Na **escala Linear**, a distância entre os ticks (pontos) na escala equivale ao mesmo valor absoluto. Já na **escala Logarítmica**, a distância entre os ticks equivale a mesma proporção.


```
#Gráfico de linha (Casos totais e mortes durante o tempo) com opção de visualização entre linear ou logarítmica.
def f(Escala):
  fig, ax = plt.subplots(figsize=(10,8))
  ax.set_title('Casos Totais e Morte Totais', size=15,)
  ax.set_xlabel('Tempo', size=12, color='grey')
  ax.set_ylabel('Numeros', size=12, color='grey')
  
  line, ax.plot(df_brasil.date, df_brasil.total_deaths, color='green')
  line, ax.plot(df_brasil.date, df_brasil.total_cases, color='red')
  ax.legend
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  ax.set_yscale(Escala)

  def major_tick(x, pos):
    if x < 1000:
      return f'{x:.0f}'
    elif x > 999999:
      return f'{x//1000000:.0f}M'
    else:
      return f'{x//1000:.0f}K'

  ax.yaxis.set_major_formatter(FuncFormatter(major_tick))
  
  fig.tight_layout()
funcao=['Linear','Log']
widgets.interact(f, Escala=funcao);


```
